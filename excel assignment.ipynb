# Question 1:What is the difference between 'Paste' and 'Paste Special' in Excel? Briefly explain with examples.
- In Excel, when we copy something and want to put it somewhere else, we usually use Paste or Paste Special. These two options work differently:

- Paste is the normal option. It pastes everything — the content, the formula (if there is one), the formatting (like colors or borders), and even comments.

- Example: If you copy a cell with a formula like =A1+B1, using Paste will paste the same formula, and it will adjust automatically depending on the new location.

- Paste Special gives you more control. You can choose what exactly you want to paste — just the value (number or text), just the formatting, just the formula, or even paste things in a different layout (like flipping rows and columns using Transpose).

- Example: If a cell has the formula =A1+B1 and shows the result "20", and you use Paste Special → Values, Excel will only paste "20" — not the formula.So, Paste is quick and simple, but Paste Special gives you more options — which is useful when you want to copy only specific parts of a cell.

# Question 2: Describe the functions and usefulness of 'Freeze Panes' and 'Split Panes' in Excel.
- Both Freeze Panes and Split Panes help you work better with large Excel sheets by improving how you see your data. But they work differently:

#Freeze Panes:

This feature "locks" certain rows or columns so they stay in place when you scroll.

Common use: Freeze the top row (with headings like Name, Age, etc.) so that even when you scroll down, the headers are still visible.

You can freeze:

Just the top row

Just the first column

Both rows and columns (by selecting a cell and using Freeze Panes)

# Split Panes:

This splits your screen into 2 or 4 sections so you can scroll and view different parts of the sheet at the same time.

It's like creating windows inside Excel.

Example: You can scroll the top half of the sheet while the bottom half stays still — useful for comparing data.

# Question 3: Explain the difference between inserting a new row and inserting a new column in Excel. Can you insert multiple rows or columns at once?

#Inserting a New Row:

Adds a horizontal line of blank cells below the selected row or rows.

Existing rows below the insertion point are shifted downward.

Example: If you insert a row at Row 5, the current Row 5 becomes Row 6.

# Inserting a New Column:

Adds a vertical line of blank cells to the right of the selected column or columns.

Existing columns to the right of the insertion point are shifted to the right.

Example: If you insert a column at Column C, the current Column C becomes Column D.

- Yes, you can insert multiple rows or columns at once by selecting the same number of existing rows or columns before using the Insert command.




# Question 4: What are logical functions in Excel? Provide examples of at least two logical functions and their applications.
- Logical functions in Excel are used to make decisions. They check if something is true or false, and based on that, Excel will show a result.

# 1. IF Function:

This is the most common logical function.

It works like this: "If something is true, do this — if not, do something else."

Example:
=IF(A2>=40, "Pass", "Fail")
→ If the value in A2 is 40 or more, Excel will show "Pass". Otherwise, it shows "Fail".

# 2. AND Function:

This checks multiple conditions at the same time.

It only returns TRUE if all conditions are true.

Example:
=AND(A1>10, B1<50)
→ Returns TRUE only if A1 is more than 10 and B1 is less than 50.


# Question 5: Discuss the purpose of 'XLOOKUP' and how it differs from the traditional 'VLOOKUP' function.
- XLOOKUP is a modern Excel function used to find and return data from a table. It’s a smarter and more flexible version of VLOOKUP.

# VLOOKUP (Old way):

Only works from left to right.

You must know the column number to return data.

Can break easily if the table changes.

# XLOOKUP (New and better way):

Can look left, right, up, or down — no direction limit.

Easier to use and doesn’t break if the table structure changes.

Can return exact matches by default.

## Question 8: Use Excel's 'Find and Replace' feature to update department names in a sample table.

### Answer:

**Step 1:** Open your Excel worksheet that contains the department names you want to update.

**Step 2:** Select the range of cells where you want to make changes (for example, the column with department names). If you want to search the entire sheet, you can skip this step.

**Step 3:** Press **Ctrl + H** on your keyboard to open the **Find and Replace** dialog box.

**Step 4:** In the **Find what** box, type the department name you want to replace.
*(Example: type “Marketing” if you want to change that)*

**Step 5:** In the **Replace with** box, type the new department name you want to use instead.
*(Example: type “Sales & Marketing”)*

**Step 6:** Click **Replace All** to replace all occurrences at once.
A message will appear telling you how many replacements were made.

**Step 7:** Click **OK**, then close the **Find and Replace** window.
