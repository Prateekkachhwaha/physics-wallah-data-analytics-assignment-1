# question-1.What is a random variable in probability theory?
- In probability theory, a random variable is a function that assigns a numerical value to each outcome in a sample space of a random experiment.

# question-2.What are the types of random variables?
- There are two main types of random variables in probability theory:

1. **Discrete Random Variable**:
   Takes on a **countable** number of distinct values (e.g., number of heads in coin tosses).

2. **Continuous Random Variable**:
   Takes on values from an **uncountable** range, typically intervals on the real number line (e.g., time, height, weight).

# question-3.What is the difference between discrete and continuous distributions.
- The **difference between discrete and continuous distributions** is given below:-

### **Discrete Distribution**:

* Deals with **countable** outcomes (e.g., 0, 1, 2, ‚Ä¶).
* Probabilities are assigned to **specific values**.
* Represented by a **probability mass function (PMF)**.
* Example: Probability of getting 2 heads in 3 coin tosses.

### **Continuous Distribution**:

* Deals with **uncountable** outcomes over an interval (e.g., any real number between 0 and 1).
* Probabilities are assigned over **intervals**, not exact values.
* Represented by a **probability density function (PDF)**.
* Example: Probability that a person‚Äôs height is between 160 cm and 170 cm.

# question-4.What are probability distribution functions (PDF)?
- A Probability Density Function (PDF) describes the likelihood of a continuous random variable taking values within a range; the total area under the curve equals 1, and exact values have zero probability.

# question-5. How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?
- difference is given below:-
  -  CDF shows the total probability up to a point.
  -  PDF is the **slope** of the CDF.
  -  CDF always increases from **0 to 1**.
  -  PDF values can be **greater than 1**, but area under curve = 1.
  -  CDF gives actual **probability**, **PDF** gives **density**.

# question-6.What is a discrete uniform distribution?
- A discrete uniform distribution is a probability distribution where all outcomes are equally likely over a finite set of values.

# question-7.What are the key properties of a Bernoulli distribution?
- Here are one-liner key properties of a Bernoulli distribution:
  - Takes values 0 (failure) or 1 (success).
  - Probability of success is $p$, failure is $1-p$.
  - Mean (expected value) is $p$.
  - Variance is $p(1-p)$.
  - Has a single parameter $p$ between 0 and 1.

# question-8.What is the binomial distribution, and how is it used in probability?
- the binomial distribution models the number of successes in a fixed number of independent trials, each with the same probability of success.

# question-9.What is the Poisson distribution and where is it applied?
- The Poisson distribution models the number of times an event occurs in a fixed interval of time or space, assuming events happen independently and at a constant average rate.
- The Poisson distribution is commonly applied in scenarios like:
  - Counting calls received at a call center per hour.
  - Number of emails arriving in your inbox in a day.
  - Counting accidents happening at a traffic intersection.
  - Number of defects in a batch of products.

# question-13.What is the Central Limit Theorem (CLT), and why is it critical in statistics?
- The Central Limit Theorem (CLT) states that the sum or average of a large number of independent, identically distributed random variables tends to follow a normal distribution, regardless of the original distribution.
- its critical becuase:-
 - It allows us to use normal distribution techniques even when the original data isn‚Äôt normal.
 - Enables confidence intervals and hypothesis tests for sample means.

# question-14. How does the Central Limit Theorem relate to the normal distribution?
- The Central Limit Theorem (CLT) explains why the normal distribution appears so often: it shows that the distribution of sample means (or sums) becomes approximately normal as the sample size grows, even if the original data isn‚Äôt normal.

# question-15.What is the application of Z statistics in hypothesis testing.
- Z-statistics are used in hypothesis testing to determine how many standard deviations a sample mean is from the population mean when the population variance is known, helping to decide whether to reject the null hypothesis.

# question-16.How do you calculate a Z-score, and what does it represent.
- Z-score tells you how far a number is from the average, measured in steps called standard deviations.
- You find it by subtracting the average from the number, then dividing by the standard deviation.

# question-17.What are point estimates and interval estimates in statistics?
- Point estimate: A single value that guesses a population parameter (like using the sample average to estimate the population average).

- Interval estimate: A range of values, called a confidence interval, that likely contains the population parameter with a certain level of confidence.

# question-18.What is the significance of confidence intervals in statistical analysis?
- Confidence intervals show a range of values where we expect the true population parameter to lie, giving a sense of how precise our estimate is and how reliable our results are.

# question-19.What is the relationship between a Z-score and a confidence interval?
- The Z-score determines the width of a confidence interval by specifying how many standard deviations to go from the sample mean to capture a certain level of confidence (like 95% or 99%).

# question-20. How are Z-scores used to compare different distributions?
- Z-scores let you standardize values from different distributions by converting them into a common scale‚Äîshowing how many standard deviations each value is from its own mean‚Äîso you can easily compare them directly.

# question-21.What are the assumptions for applying the Central Limit Theorem?
- The main assumptions for applying the Central Limit Theorem (CLT) are:
 - Independence: The samples must be independent of each other.
 - Identically Distributed: The samples should come from the same distribution (with the same mean and variance).

# question-22.What is the concept of expected value in a probability distribution?
- The expected value of a probability distribution is the long-run average or mean value you expect if you repeat the random experiment many times.

# question-23. How does a probability distribution relate to the expected outcome of a random variable?
- A probability distribution shows all possible outcomes of a random variable and their chances, and the expected outcome (or expected value) is the weighted average of these outcomes based on their probabilities.
