question 14. How does the “in” keyword work differently for lists and dictionaries ?

- **For lists**: The `in` keyword checks if a specific **value** exists in the list. It goes through each item in the list to see if it matches.

  **Example**:
  ```python
  my_list = [1, 2, 3, 4]
  print(3 in my_list)  # Output: True
  ```

- **For dictionaries**: The `in` keyword checks if a **key** exists in the dictionary, not the value.

  **Example**:
  ```python
  my_dict = {"apple": 10, "banana": 5}
  print("apple" in my_dict)  # Output: True
  ```

question 19. Why are dictionaries considered unordered, and how does that affect data retrieval?
- Dictionaries in Python are considered unordered because they do not maintain the order of the keys based on their insertion sequence (at least, not in the way a list or tuple would). This is a key distinction that comes from the way dictionaries are implemented internally.

### Why are dictionaries unordered?
Dictionaries in Python are implemented using a **hash table**. Here’s a simple breakdown of how this works:

1. **Hashing**: When you add a key-value pair to a dictionary, the key is passed through a hashing function that computes an index (or hash) in the underlying array.
   
2. **No guaranteed order**: The hash table doesn’t inherently track the order of insertion. It just places each key-value pair at a specific index based on the hash of the key,

3. **Efficiency**: The primary advantage of using a hash table (and why dictionaries are implemented this way) is speed. Lookups, inserts, and deletions are typically done in constant time, O(1), which is very efficient.

### Does this affect data retrieval?
While dictionaries are unordered, this doesn’t affect the ability to retrieve data **by key**. You can still access any value in a dictionary by referencing its key, and the retrieval will be fast (O(1) on average).





question 20.  Explain the difference between a list and a dictionary in terms of data retrieval.

 **List**:
- **Data Structure**: A list is an **ordered collection** of items (e.g., numbers, strings).
- **Data Retrieval**: You retrieve items by their **index** (position in the list).
  - **Example**: `my_list[2]` will get the 3rd item in the list.
- **Time Complexity**: **O(1)** to retrieve an item by index.

**Dictionary**:
- **Data Structure**: A dictionary is an **unordered collection** of **key-value pairs**.
- **Data Retrieval**: You retrieve items by their **key**, not by index.
  - **Example**: `my_dict['apple']` will get the value associated with the key `'apple'`.
- **Time Complexity**: **O(1)** to retrieve an item by key.

**Key Differences**:
- **List**: Retrieve by **index** (position).
- **Dictionary**: Retrieve by **key** (name or identifier).
