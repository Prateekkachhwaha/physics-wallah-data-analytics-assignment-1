# question-1.What is hypothesis testing in statistics?
- Hypothesis testing in statistics is a method used to make decisions or inferences about a population based on a sample of data. It involves testing an assumption about a population parameter.

# question-2.What is the null hypothesis, and how does it differ from the alternative hypothesis?
- The null hypothesis is a statistical assumption that says that  there is no effect, no difference, or no relationship in a population — it acts likee an  default or starting point for testing.
###the difference:-

   - The null hypothesis says  there is no effect or difference, while the alternative hypothesis says that there is an effect or difference.

# question-3.What is the significance level in hypothesis testing, and why is it important?
- The significance level  is basically the cutoff you choose to decide what counts as strong enough evidence to reject the default assumption.
- It's important because it helps you decide when your test results are convincing enough to stop believing the status quo  and accept that something different is likely going on.

# question-4.What does a P-value represent in hypothesis testing?
- The p-value in hypothesis testing tells you how likely your results are, assuming the null hypothesis is true.

# question-5.How do you interpret the P-value in hypothesis testing?
- The p-value shows how likely your results are if the null hypothesis is true. A small p-value (≤ 0.05) means you reject the null; a large p-value means you don’t.

# question-6.What are Type 1 and Type 2 errors in hypothesis testing?
- Type 1 error: Rejecting the null hypothesis when it’s actually true (a false alarm).
- Type 2 error: Failing to reject the null hypothesis when it’s actually false (missing a real effect).

# question-7.What is the difference between a one-tailed and a two-tailed test in hypothesis testing?
- One-tailed test checks for an effect in one specific direction (either greater than or less than).
- Two-tailed test checks for an effect in both directions (either greater than or less than).

# question-8.What is the Z-test, and when is it used in hypothesis testing?
- A Z-test is a statistical test used to compare a sample mean to a population mean when the population standard deviation is known and the sample size is large (usually 30 or more). It helps determine if there’s a significant difference between them.

# question-9.How do you calculate the Z-score, and what does it represent in hypothesis testing?
- we calculate z-score by using this formula:-
  ### sample mean-population mean / population standard deviation/sample size

- It represents how many standard deviations your sample mean is away from the population mean — helping you see if the difference is statistically significant.

# question-10. What is the T-distribution, and when should it be used instead of the normal distribution?
- The T-distribution is like the normal (bell curve) distribution but with thicker tails, which means it accounts for more uncertainty.
  - you use the T-distribution instead of the normal distribution when:
     - Your sample size is small (usually less than 30)

# question-11.What is the difference between a Z-test and a T-test?
- Z-test: Used when the population standard deviation is known and the sample size is large.
- T-test: Used when the population standard deviation is unknown and the sample size is small.

# question-12.What is the T-test, and how is it used in hypothesis testing?
- T-test: A test used to compare means when the population standard deviation is unknown and the sample size is small.
   -  It helps decide if the difference between sample and population (or between two samples) is statistically significant.

# question-13.What is the relationship between Z-test and T-test in hypothesis testing?
- The Z-test and T-test both compare means, but the Z-test is used when the population standard deviation is known and the sample is large, while the T-test is used when the standard deviation is unknown and the sample is small.

# question-14.What is a confidence interval, and how is it used to interpret statistical results?
- A confidence interval is a range of values that likely contains the true population parameter, giving you an idea of the estimate’s precision.
- It’s used to show how confident you can be about your sample estimate

# question-15.What is the margin of error, and how does it affect the confidence interval?
- The margin of error is the amount added and subtracted from a sample estimate to create a confidence interval.A larger margin of error means a wider confidence interval, showing less precision, while a smaller margin means a narrower, more precise interval.

# question-16.>How is Bayes' Theorem used in statistics, and what is its significance?
- Bayes' Theorem is used in statistics to update the probability of a hypothesis based on new evidence.
- Its significance lies in combining prior knowledge with current data to make better, more informed decisions or predictions.

# question-17.What is the Chi-square distribution, and when is it used?
- The Chi-square distribution is a probability distribution used mainly for tests involving categorical data.
- It’s commonly used in Chi-square tests to check if observed data fits expected patterns, like testing independence or goodness of fit.

# question-18.What is the Chi-square goodness of fit test, and how is it applied?
- The Chi-square goodness of fit test checks whether observed data matches an expected distribution.
- You apply it by comparing your observed frequencies to expected frequencies and calculating a Chi-square statistic to see if any differences are due to chance or indicate a real mismatch.

# question-19.What is the F-distribution, and when is it used in hypothesis testing?
- The F-distributionis a probability distribution used to compare variances between two groups.
- It’s commonly used in ANOVA tests and F-tests to determine if group variances are significantly different.

# question-20.What is an ANOVA test, and what are its assumptions?
- ANOVA (Analysis of Variance) is a test used to compare the means of three or more groups to see if at least one group mean is different from the others.
###Assumptions of ANOVA:
     - The groups are independent.
     - The data in each group are normally distributed.

# question-21.What are the different types of ANOVA tests?
- there are different types of ANOVA tests which are stated below:-
   - One-way ANOVA: Compares means across one factor with three or more groups.
   - Two-way ANOVA: Compares means across two factors and can check for interaction effects between them.
   - Repeated Measures ANOVA: Compares means when the same subjects are measured under different conditions or times.

# question-22.What is the F-test, and how does it relate to hypothesis testing?
- The F-test compares the variances of two or more groups to see if they are significantly different.
- In hypothesis testing, it’s often used to determine if group means differ by analyzing variance (like in ANOVA). A large F-value suggests at least one group mean is different.

# question-10. Implement the Central Limit Theorem (CLT) using Python for a non-normal distribution.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def demonstrate_clt(population_size=100000, sample_size=50, num_samples=1000, rate=1.0):
    population = np.random.exponential(scale=1/rate, size=population_size)
    sample_means = []
    for _ in range(num_samples):
        sample = np.random.choice(population, size=sample_size, replace=False)
        sample_means.append(np.mean(sample))
    plt.figure(figsize=(14, 6))
    plt.subplot(1, 2, 1)
    sns.histplot(population, bins=100, kde=True, color='skyblue')
    plt.title('Original Exponential Distribution')
    plt.xlabel('Value')
    plt.ylabel('Frequency')
    plt.subplot(1, 2, 2)
    sns.histplot(sample_means, bins=30, kde=True, color='salmon')
    plt.title(f'Distribution of Sample Means\n(Sample Size = {sample_size}, Samples = {num_samples})')
    plt.xlabel('Sample Mean')
    plt.ylabel('Frequency')

    plt.tight_layout()
    plt.show()

demonstrate_clt()
```

```python
from scipy.stats import norm

def calculate_z_score(x, mean, std):
    z = (x - mean) / std
    return z

def compare_to_standard_normal(x, mean, std):
    z = calculate_z_score(x, mean, std)
    prob = norm.cdf(z)
    print(f"Data point: {x}")
    print(f"Z-score: {z:.2f}")
    print(f"Probability (P(Z < {z:.2f})): {prob:.4f}")
compare_to_standard_normal(x=75, mean=70, std=5)
```

# question-15.Implement hypothesis testing using Z-statistics for a sample dataset.

```python
import numpy as np
from scipy.stats import norm

def z_test(sample_data, population_mean, population_std, alpha=0.05):
    sample_mean = np.mean(sample_data)
    sample_size = len(sample_data)

    z = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

    p_value = 2 * (1 - norm.cdf(abs(z)))

    print(f"Sample Mean: {sample_mean:.2f}")
    print(f"Z-statistic: {z:.2f}")
    print(f"P-value: {p_value:.4f}")

    if p_value < alpha:
        print("Reject the null hypothesis (significant difference).")
    else:
        print("Fail to reject the null hypothesis (no significant difference).")

sample = [105, 110, 98, 100, 102, 108, 107, 101, 103, 106]

z_test(sample_data=sample, population_mean=100, population_std=5)
```

# question-16. Create a confidence interval for a dataset using Python and interpret the result.

```python
import numpy as np
from scipy import stats

def confidence_interval(data, confidence=0.95):

    n = len(data)
    mean = np.mean(data)
    std_err = stats.sem(data)
    h = std_err * stats.t.ppf((1 + confidence) / 2, n - 1)

    lower = mean - h
    upper = mean + h

    print(f"Sample Mean: {mean:.2f}")
    print(f"{int(confidence * 100)}% Confidence Interval: ({lower:.2f}, {upper:.2f})")
    print(f"We are {int(confidence * 100)}% confident that the true population mean lies between {lower:.2f} and {upper:.2f}.")

    return (lower, upper)
sample_data = [12, 15, 14, 10, 13, 15, 14, 16, 15, 14]
confidence_interval(sample_data, confidence=0.95)
```

# question-17. Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean.

```python
import numpy as np
from scipy import stats

def generate_data_and_confidence_interval(mean=50, std=5, size=100, confidence=0.95):

    data = np.random.normal(loc=mean, scale=std, size=size)
    n = len(data)
    sample_mean = np.mean(data)
    std_err = stats.sem(data)

    h = std_err * stats.t.ppf((1 + confidence) / 2, n - 1)

    lower = sample_mean - h
    upper = sample_mean + h

    print(f"Sample Mean: {sample_mean:.2f}")
    print(f"{int(confidence*100)}% Confidence Interval: ({lower:.2f}, {upper:.2f})")
    print(f"Interpretation: We are {int(confidence*100)}% confident that the true population mean lies between {lower:.2f} and {upper:.2f}.")

    return data, (lower, upper)
generate_data_and_confidence_interval(mean=50, std=5, size=100, confidence=0.95)
```

# question-1.Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results.

```python
from scipy.stats import norm
import numpy as np
sample = [100, 102, 98, 97, 101, 99, 100, 103, 95, 96]
pop_mean = 100
pop_std = 3
sample_mean = np.mean(sample)
n = len(sample)
z = (sample_mean - pop_mean) / (pop_std / np.sqrt(n))
p = 2 * (1 - norm.cdf(abs(z)))
print(f"Z = {z:.2f}, p = {p:.4f}")
if p < 0.05:
    print("Result: Significant difference (reject H0)")
else:
    print("Result: No significant difference (fail to reject H0)")
```

# question-4. Perform a two-tailed Z-test using Python and visualize the decision region on a plot.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

sample = [102, 98, 101, 105, 99, 97, 103, 100, 96, 104]
pop_mean = 100
pop_std = 5
alpha = 0.05

sample_mean = np.mean(sample)
n = len(sample)
z = (sample_mean - pop_mean) / (pop_std / np.sqrt(n))
p_value = 2 * (1 - norm.cdf(abs(z)))
z_critical = norm.ppf(1 - alpha/2)

print(f"Sample Mean = {sample_mean:.2f}")
print(f"Z-score = {z:.2f}")
print(f"P-value = {p_value:.4f}")
print(f"Critical Z-values = ±{z_critical:.2f}")
if abs(z) > z_critical:
    print("Result: Significant difference (reject H0)")
else:
    print("Result: No significant difference (fail to reject H0)")

x = np.linspace(-4, 4, 1000)
y = norm.pdf(x, 0, 1)

plt.figure(figsize=(10, 5))
plt.plot(x, y, label='Standard Normal Distribution')
plt.fill_between(x, y, where=(x < -z_critical) | (x > z_critical), color='red', alpha=0.3, label='Rejection Region')
plt.axvline(z, color='blue', linestyle='--', label=f'Sample Z = {z:.2f}')
plt.axvline(-z_critical, color='red', linestyle='--')
plt.axvline(z_critical, color='red', linestyle='--')
plt.title('Two-Tailed Z-Test: Decision Region')
plt.xlabel('Z')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```

# question-5.Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def visualize_type1_type2_errors(mu_0, mu_1, sigma, n, alpha=0.05):
    se = sigma / np.sqrt(n)

    z_critical = norm.ppf(1 - alpha)  # one-tailed test (right tail)
    x_critical = mu_0 + z_critical * se

    beta = norm.cdf((x_critical - mu_1) / se)
    power = 1 - beta

    x = np.linspace(mu_0 - 4*se, mu_1 + 4*se, 1000)
    h0_dist = norm.pdf(x, mu_0, se)
    h1_dist = norm.pdf(x, mu_1, se)

    plt.figure(figsize=(10, 5))
    plt.plot(x, h0_dist, label='H0 Distribution (μ = {:.1f})'.format(mu_0), color='blue')
    plt.plot(x, h1_dist, label='H1 Distribution (μ = {:.1f})'.format(mu_1), color='green')

    plt.fill_between(x, 0, h0_dist, where=(x > x_critical), color='red', alpha=0.3, label='Type I Error (α)')
    plt.fill_between(x, 0, h1_dist, where=(x < x_critical), color='orange', alpha=0.3, label='Type II Error (β)')

    plt.axvline(x_critical, color='black', linestyle='--', label='Critical Value')

    plt.title('Type I and Type II Errors in Hypothesis Testing')
    plt.xlabel('Sample Mean')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.show()

    print(f"Critical Value (Sample Mean) = {x_critical:.2f}")
    print(f"Type I Error (α) = {alpha:.2f}")
    print(f"Type II Error (β) = {beta:.4f}")
    print(f"Power of the Test = {power:.4f}")

visualize_type1_type2_errors(mu_0=100, mu_1=103, sigma=10, n=30, alpha=0.05)
```

# question-7.Perform a paired sample T-test using Python and visualize the comparison results.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import ttest_rel

# Example paired data (e.g., before and after treatment)
before = [85, 90, 88, 75, 95, 92, 80, 77, 91, 89]
after  = [87, 93, 90, 78, 97, 94, 82, 79, 93, 90]

# Perform paired t-test
t_stat, p_value = ttest_rel(before, after)

# Output results
print(f"T-statistic = {t_stat:.2f}")
print(f"P-value = {p_value:.4f}")
if p_value < 0.05:
    print("Result: Significant difference (reject H0)")
else:
    print("Result: No significant difference (fail to reject H0)")

# Visualization
plt.figure(figsize=(10, 5))

# Line plot: connecting paired samples
for i in range(len(before)):
    plt.plot([0, 1], [before[i], after[i]], marker='o', color='gray', alpha=0.7)

# Boxplot: side-by-side
plt.boxplot([before, after], labels=['Before', 'After'])
plt.title('Paired Sample Comparison')
plt.ylabel('Scores')
plt.grid(True)
plt.show()
```

# question-9.Write a Python function to calculate the confidence interval for a sample mean and explain its significance.

```python
import numpy as np
from scipy.stats import norm, t

def confidence_interval(data, confidence=0.95, known_std=None):
    n = len(data)
    mean = np.mean(data)

    if known_std is not None:
        # Use Z-distribution if population std dev known
        std_err = known_std / np.sqrt(n)
        z = norm.ppf((1 + confidence) / 2)
        margin = z * std_err
    else:
        # Use T-distribution if population std dev unknown
        sample_std = np.std(data, ddof=1)
        std_err = sample_std / np.sqrt(n)
        t_score = t.ppf((1 + confidence) / 2, df=n-1)
        margin = t_score * std_err

    lower = mean - margin
    upper = mean + margin
    return (lower, upper)

# Example usage
sample_data = [100, 102, 98, 101, 99, 97, 103, 100, 96, 104]
ci_lower, ci_upper = confidence_interval(sample_data, confidence=0.95)
print(f"95% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")
```

# question-10. Write a Python program to calculate the margin of error for a given confidence level using sample data.

```python
import numpy as np
from scipy.stats import norm, t

def margin_of_error(data, confidence=0.95, known_std=None):
    n = len(data)
    if known_std is not None:
        std_err = known_std / np.sqrt(n)
        z = norm.ppf((1 + confidence) / 2)
        margin = z * std_err
    else:
        sample_std = np.std(data, ddof=1)
        std_err = sample_std / np.sqrt(n)
        t_score = t.ppf((1 + confidence) / 2, df=n-1)
        margin = t_score * std_err
    return margin
sample_data = [100, 102, 98, 101, 99, 97, 103, 100, 96, 104]
moe = margin_of_error(sample_data, confidence=0.95)
print(f"Margin of Error (95% confidence): ±{moe:.2f}")
```

# question-12. Perform a Chi-square test for independence between two categorical variables in Python.

```python
import numpy as np
from scipy.stats import chi2_contingency

data = np.array([[30, 10],
                 [20, 40]])

chi2, p, dof, expected = chi2_contingency(data)

print(f"Chi-square statistic = {chi2:.3f}")
print(f"Degrees of freedom = {dof}")
print(f"P-value = {p:.4f}")

if p < 0.05:
    print("Result: Variables are dependent (reject null hypothesis)")
else:
    print("Result: Variables are independent (fail to reject null hypothesis)")
```

# question-13.Write a Python program to calculate the expected frequencies for a Chi-square test based on observed data.

```python
import numpy as np

observed = np.array([[30, 10],
                     [20, 40]])

row_sums = observed.sum(axis=1).reshape(-1, 1)
col_sums = observed.sum(axis=0).reshape(1, -1)
total = observed.sum()

expected = (row_sums @ col_sums) / total

print("Expected Frequencies:")
print(expected)

```

# question-14.Perform a goodness-of-fit test using Python to compare the observed data to an expected distribution.

```python
import numpy as np
from scipy.stats import chisquare

observed = np.array([50, 30, 20])
expected = np.array([40, 40, 20])

chi2_stat, p_value = chisquare(f_obs=observed, f_exp=expected)

print(f"Chi-square statistic = {chi2_stat:.3f}")
print(f"P-value = {p_value:.4f}")

if p_value < 0.05:
    print("Result: Observed data significantly differs from expected distribution (reject H0)")
else:
    print("Result: No significant difference between observed and expected (fail to reject H0)")
```
